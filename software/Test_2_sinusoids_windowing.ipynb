```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from scipy.signal import get_window
sys.path.append('../software/models/')
import dftModel as DFT

fs = 44100
M = 30 # number of samples in input signal, same as window size

f1 = 5000.0
f2 = 9000.0
x = np.cos(2 * np.pi * f1 * np.arange(M) / float(fs)) # + np.cos(2 * np.pi * f2 * np.arange(M) / float(fs))

w = get_window('blackman', M)
N = 512 # fftSize (power of 2)
mX, pX = DFT.dftAnal(x, w, N)

# x axis; from 0 to half fs, with steps of fs/N
plt.plot(np.arange(0, fs/2, fs/float(N)), mX[:int(N/2)]-max(mX))
plt.show()
```
